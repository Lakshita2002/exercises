```python
import requests
page = requests.get('http://www.gibbon.se/Retailer/Map.aspx?SectionId=832')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
countries = soup.find_all('div', attrs={'class':'dropdown'})
countries
```

```python
countries[0].text
```

```python
countries[0].text.split('\n')
```

```python
country_list = countries[0].text.split('\n')
country_list = country_list[2:-2]
```

```python
country_list
```

```python
len(country_list)
```

```python
import pandas as pd
df = pd.DataFrame(country_list, columns=['countries'])
```
