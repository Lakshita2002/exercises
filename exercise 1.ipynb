```python
import requests
page = requests.get('https://support.sas.com/en/papers/proceedings-archive/sugi2005.html')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
papers = soup.find_all('cite')
papers
```

```python
papers[0].text
```

```python
dataset = []
for data in papers:
    name = data.text
    dataset.append((name))
```

```python
len(dataset)
```

```python
dataset[0:5]
```

```python
import pandas as pd
df = pd.DataFrame(dataset, columns=['name'])
```
